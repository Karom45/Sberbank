```python
lst = []
new_lst = []
pattern = [' Part of ','\n' , ' ','']
def condition(i):
    return "img" not in [ch.name for ch in i.children] and len(i.get('href').split('/')[-1].split('#')) == 1
for TR in root.find_all('tr')[1:2]:
    for TD in TR.find_all('td')[1:]:
        for child in TD:
            if child.name == 'br':
                lst.append(new_lst)
                new_lst = []
            else:
                new_lst.append(child)
        else:
            lst.append(new_lst)
# print(lst[0])                               
for num , i in enumerate(lst):
    for num1 , item in enumerate(i):
        try:
#             print(item)
            item_hr = item.find("a")  
#             print(item_hr , item)
            if item_hr == None:
                item_hr = item
            if isinstance(item_hr, int):
                continue
            if not condition(item_hr):
                if item_hr.get('title') and "img" not in [ch.name for ch in item.children]:
                    title = re.sub('[ ]',"_" ,item_hr.get("title"))
                    lst[num].append(f'{title}_title')
                del lst[num][num1]
            else:
#                 print(item_hr.get('href').split('/')[-1])
                lst[num].append(item_hr.get('href').split('/')[-1])
                del lst[num][num1]
        except Exception as e:
            print(f"{item}, ошибка :{e}")
            if len(item.split('–'))==1 :
                del lst[num][num1]
                
for i,item in enumerate(lst):
    del_lst = []
    for j,val in enumerate(item):
        if lst[i][j] in pattern:
            del_lst.append(val)
            continue
        lst[i][j] = re.sub('[() \n]',"",lst[i][j])
        if len(lst[i][j].split(';')) >1:
            del_lst.append(lst[i][j])
            lst[i].extend(lst[i][j].split(';'))
    for val in del_lst:
        lst[i].remove(val)


lst[1]
```

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_predecessors_of_sovereign_states_in_Europe')
soup = BeautifulSoup(r.text , "lxml")
root = soup.find_all('table')[3]
# print(root)
dic = defaultdict(list)
def condition(i):
    return "img" not in [ch.name for ch in i.children] and len(i.get('href').split('/')[-1].split('#')) == 1
def not_image(tag):
        return not tag.name!="img"

for item in root.find_all('tr')[1:]:
    for i in item.find_all('a')[1:]:
        # print(i)
        if condition(i):
        # print(i,i.previous.previous.previous)
            date = i.next.next
            if isinstance(date, Tag) or date == ')':
                date = i.previous.previous.previous
            if isinstance(date, Tag):
                date = ""
        #print(date)
            date = re.sub('[() \n]',"",date).split(';')
            for part in date:
                try:
                    dic[item.a.get('href').split('/')[-1]].append({i.get('href').split('/')[-1].split('#')[0] : {
                                    "begin" : part.split('–')[0] , "end" :  part.split('–')[1]}})
                except:
                    dic[item.a.get('href').split('/')[-1]].append({i.get('href').split('/')[-1].split('#')[0] : {
                                    "begin" : 'нет' , "end" : 'нет'}})
dic
```
